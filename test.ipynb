```python
import pandas as pd
import matplotlib.pyplot as plt
points = pd.read_csv('dataData.csv')
points = points.values
fig, ax = plt.subplots(subplot_kw=dict(projection='3d'))
ax.scatter(*points.T)
ax.set_aspect('equal')
plt.show()
```

```python
from sklearn.manifold import TSNE
from umap import UMAP
from trimap import TRIMAP
from pacmap import PaCMAP

def TSNE_reduce(data):
    tsne = TSNE()
    return tsne.fit_transform(data)

def Umap_reduce(data):
    umap = UMAP()
    return umap.fit_transform(data)

def Trimap_reduce(data):
    trimap = TRIMAP()
    return trimap.fit_transform(data)

def Pacmap_reduce(data):
    pacmap = PaCMAP()
    return pacmap.fit_transform(data)

def IVHD_reduce(data):
    ...

def embeded_visualization(data, title):
    # rozbicie punktw na wsprzdne x, y, z
    x = data[:, 0]
    y = data[:, 1]

    fig, ax = plt.subplots()
    # wygenerowanie wykresu punktw
    ax.scatter(x, y, c='green')

    plt.title(f"{title} Visualization")
    plt.xlabel("Dimension 1")
    plt.ylabel("Dimension 2")

    # wywietlenie wykresu
    plt.show()

def embeded_visualization_two_clases(data, N, title):
    # rozbicie punktw na wsprzdne x, y, z
    x = data[:N, 0]
    y = data[:N, 1]

    x1 = data[N:2*N, 0]
    y1 = data[N:2*N, 1]

    fig, ax = plt.subplots()
    # wygenerowanie wykresu punktw
    ax.scatter(x, y, c='green')
    ax.scatter(x1, y1, c='blue')

    plt.title(f"{title} Visualization")
    plt.xlabel("Dimension 1")
    plt.ylabel("Dimension 2")

    # wywietlenie wykresu
    plt.show()

def embeded_visualization_three_clases(data, N, title):
    # rozbicie punktw na wsprzdne x, y, z
    x = data[:N, 0]
    y = data[:N, 1]

    x1 = data[N:2*N, 0]
    y1 = data[N:2*N, 1]

    x2 = data[2*N:, 0]
    y2 = data[2*N:, 1]

    fig, ax = plt.subplots()
    # wygenerowanie wykresu punktw
    ax.scatter(x, y, c='green')
    ax.scatter(x1, y1, c='blue')
    ax.scatter(x2, y2, c='red')

    plt.title(f"{title} Visualization")
    plt.xlabel("Dimension 1")
    plt.ylabel("Dimension 2")

    # wywietlenie wykresu
    plt.show()
```

```python
reduced = TSNE_reduce(points)
```

```python
embeded_visualization(reduced, 'plane_tsne')
```

```python
tsne = TSNE(perplexity=30)
transformed = tsne.fit_transform(points)
embeded_visualization(transformed, 't-sne')
```

```python
reduced = Umap_reduce(points)
embeded_visualization(reduced, 'umap')
```
